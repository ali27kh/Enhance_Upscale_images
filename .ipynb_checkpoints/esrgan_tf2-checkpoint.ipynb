```python
import os
import time
from PIL import Image
import cv2
import tensorflow as tf
import tensorflow_hub as hub
import numpy as np
import matplotlib.pyplot as plt
```

```python
os.environ["TFHUB_DOWNLOAD_PROGRESS"] = "True"
```

```python
def preprocess_image(image_path):
    """Prépare l'image pour l'entrée dans le modèle en supprimant le canal alpha et en ajustant la taille."""
    hr_image = tf.image.decode_image(tf.io.read_file(image_path))
    if hr_image.shape[-1] == 4:  # Retirer le canal alpha si PNG
        hr_image = hr_image[...,:-1]
    hr_size = (tf.convert_to_tensor(hr_image.shape[:-1]) // 4) * 4
    hr_image = tf.image.crop_to_bounding_box(hr_image, 0, 0, hr_size[0], hr_size[1])
    return tf.expand_dims(tf.cast(hr_image, tf.float32), 0)

def plot_image(image, title=""):
    """Affiche une image avec un titre."""
    image = np.asarray(image)
    image = tf.clip_by_value(image, 0, 255)
    image = Image.fromarray(tf.cast(image, tf.uint8).numpy())
    plt.imshow(image)
    plt.axis("off")
    plt.title(title)

def save_image(image, filename):
    """Enregistre une image après l'avoir convertie en format acceptable pour le disque."""
    if not isinstance(image, Image.Image):
        image = tf.clip_by_value(image, 0, 255)
        image = Image.fromarray(tf.cast(image, tf.uint8).numpy())
    image.save(f"{filename}.jpg")
    print(f"Saved as {filename}.jpg")

def denoise_image(image):
    """Applique le débruitage sur l'image."""
    return cv2.fastNlMeansDenoisingColored(image, None, 10, 10, 7, 21)
```

```python
model = hub.load("esrgan_model_directory")

# Chargement de l'image à traiter
img_path = 'test1.png'
hr_image = preprocess_image(img_path)
```

```python
plot_image(tf.squeeze(hr_image), title="Original Image")
```

```python
start = time.time()
fake_image = model(hr_image)
print(f"Time Taken: {time.time() - start:.6f} seconds")
```

```python
plot_image(tf.squeeze(fake_image), title="Super Resolution")
```

```python
save_image(tf.squeeze(fake_image), filename="test1_esrgantf2")
```
