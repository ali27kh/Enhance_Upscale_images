```python
import os
import cv2
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from PIL import Image
from cv2 import dnn_superres
from skimage.metrics import structural_similarity as ssim
```

```python
# list models
models = [
    ('edsr', 4, 'EDSR_x4.pb'),
    ('lapsrn', 4, 'LapSRN_x4.pb'),
    ('espcn', 4, 'ESPCN_x4.pb'),
    ('fsrcnn', 4, 'FSRCNN_x4.pb'),
]
```

```python
# Apply a super-r√©solution
def apply_sr_model(model_name, scale, model_path, image):
    sr = dnn_superres.DnnSuperResImpl_create()
    sr.readModel(model_path)
    sr.setModel(model_name, scale)
    upscaled = sr.upsample(image)
    return upscaled

# Calculate PSNR
def calculate_psnr(original, upscaled):
    if original.shape != upscaled.shape:
        upscaled = cv2.resize(upscaled, (original.shape[1], original.shape[0]), interpolation=cv2.INTER_CUBIC)
    return cv2.PSNR(original, upscaled)

# Calculate SSIM
def calculate_ssim(original, upscaled):
    if original.shape[:2] != upscaled.shape[:2]:
        upscaled = cv2.resize(upscaled, (original.shape[1], original.shape[0]), interpolation=cv2.INTER_CUBIC)
    original_gray = cv2.cvtColor(original, cv2.COLOR_BGR2GRAY)
    upscaled_gray = cv2.cvtColor(upscaled, cv2.COLOR_BGR2GRAY)
    ssim_index, _ = ssim(original_gray, upscaled_gray, full=True)
    return ssim_index
# Plot image
def plot_image(image, title=""):
    """Displays an image with a title."""
    image = np.asarray(image)
    image = tf.clip_by_value(image, 0, 255)
    image = Image.fromarray(tf.cast(image, tf.uint8).numpy())
    plt.imshow(image)
    plt.axis("off")
    plt.title(title)
    plt.show()
    
# Test
def test_super_resolution(image_path):
    output_dir = 'output_results'
    os.makedirs(output_dir, exist_ok=True)
    image = cv2.imread(image_path)
    
    if image is None:
        print(f"Image not found or invalid format: {image_path}")
        return
    
    results = []
    best_image = None

    for model_name, scale, model_path in models:
        try:
            if not os.path.exists(model_path):
                print(f'Model file not found: {model_path}')
                continue
            
            upscaled_image = apply_sr_model(model_name, scale, model_path, image)
            psnr_value = calculate_psnr(image, upscaled_image)
            ssim_value = calculate_ssim(image, upscaled_image)
            results.append((model_name, scale, psnr_value, ssim_value, upscaled_image))
            
            print(f'{model_name} - x{scale} | PSNR: {psnr_value:.2f} | SSIM: {ssim_value:.4f}')
            output_path = os.path.join(output_dir, f'{model_name}_x{scale}.png')
            cv2.imwrite(output_path, upscaled_image)
        
        except Exception as e:
            print(f'Error applying {model_name} - x{scale}: {e}')

    if results:
        best_model = max(results, key=lambda x: (x[2], x[3]))
        print(f'\nBest Model: {best_model[0]} - x{best_model[1]} | PSNR: {best_model[2]:.2f} | SSIM: {best_model[3]:.4f}')
        
        # Display the best image
        plot_image(best_model[4], title=f"Best Model: {best_model[0]} - x{best_model[1]}")
    else:
        print('No valid results found.')
```

```python
test_super_resolution('test1.png')
```
